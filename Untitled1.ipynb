```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import imageio

image_folder = '/content/drive/MyDrive/kitti/testing/image_2'
calib_folder = '/content/drive/MyDrive/kitti/testing/calib'
result_folder = '/content/drive/MyDrive/mmdetection3d/results/pred_instances_3d'
output_folder = '/content/drive/MyDrive/mmdetection3d/visualized_frames'
output_video_path = '/content/drive/MyDrive/mmdetection3d/kitti_detections_video.mp4'

os.makedirs(output_folder, exist_ok=True)

def compute_box_3d(dim, location, rotation_y):
    h, w, l = dim
    x, y, z = location

    x_corners = [l/2, l/2, -l/2, -l/2, l/2, l/2, -l/2, -l/2]
    y_corners = [0, 0, 0, 0, -h, -h, -h, -h]
    z_corners = [w/2, -w/2, -w/2, w/2, w/2, -w/2, -w/2, w/2]

    corners_3d = np.vstack([x_corners, y_corners, z_corners])
    R = np.array([
        [np.cos(rotation_y), 0, np.sin(rotation_y)],
        [0, 1, 0],
        [-np.sin(rotation_y), 0, np.cos(rotation_y)]
    ])
    corners_3d = R @ corners_3d
    corners_3d = corners_3d + np.array([[x], [y], [z]])
    return corners_3d

def project_to_image(pts_3d, P):
    pts_3d_hom = np.vstack((pts_3d, np.ones((1, pts_3d.shape[1]))))
    pts_2d = P @ pts_3d_hom
    pts_2d[:2, :] /= pts_2d[2, :]
    return pts_2d[:2, :]

def draw_projected_box3d(img, qs, color=(0,255,0), thickness=2):
    qs = qs.astype(int).T
    for k in range(0,4):
        i,j=k,(k+1)%4
        cv2.line(img, tuple(qs[i]), tuple(qs[j]), color, thickness)
        cv2.line(img, tuple(qs[i+4]), tuple(qs[j+4]), color, thickness)
        cv2.line(img, tuple(qs[i]), tuple(qs[i+4]), color, thickness)
    return img

frame_ids = sorted(f[:-4] for f in os.listdir(result_folder) if f.endswith('.txt'))

for frame_id in tqdm(frame_ids):
    image_path = os.path.join(image_folder, f"{frame_id}.png")
    calib_path = os.path.join(calib_folder, f"{frame_id}.txt")
    result_path = os.path.join(result_folder, f"{frame_id}.txt")

    if not os.path.exists(image_path) or not os.path.exists(calib_path):
        continue

    img = cv2.imread(image_path)

    with open(calib_path, 'r') as f:
        lines = f.readlines()
    P2 = np.array([float(x) for x in lines[2].split()[1:]]).reshape(3, 4)

    with open(result_path, 'r') as f:
        lines = f.readlines()

    for line in lines:
        data = line.strip().split()
        cls = data[0]
        if cls == 'DontCare':
            continue
        trunc, occl, alpha = float(data[1]), int(data[2]), float(data[3])
        xmin, ymin, xmax, ymax = map(float, data[4:8])
        h, w, l = map(float, data[8:11])
        x, y, z = map(float, data[11:14])
        ry = float(data[14])

        box3d_pts_3d = compute_box_3d((h, w, l), (x, y, z), ry)
        box3d_pts_2d = project_to_image(box3d_pts_3d, P2)

        img = draw_projected_box3d(img, box3d_pts_2d, color=(0,255,0))

        cv2.putText(img, cls, (int(xmin), int(ymin)-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,0), 1)

    output_image_path = os.path.join(output_folder, f"{frame_id}.png")
    cv2.imwrite(output_image_path, img)

images = [cv2.imread(os.path.join(output_folder, f)) for f in sorted(os.listdir(output_folder)) if f.endswith('.png')]
height, width, _ = images[0].shape
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), 5, (width, height))

for img in images:
    out.write(img)
out.release()

print(f"🎥 Video saved to {output_video_path}")


```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import os
import imageio.v2 as imageio  # use v2 to avoid deprecated warnings
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas

# KITTI helper: load point cloud from .bin file
def load_velodyne_bin(bin_path):
    return np.fromfile(bin_path, dtype=np.float32).reshape(-1, 4)[:, :3]

# KITTI helper: load label from .txt file
def load_kitti_labels(label_path):
    boxes = []
    with open(label_path, 'r') as f:
        for line in f.readlines():
            obj = line.strip().split(' ')
            if obj[0] not in ['Car', 'Pedestrian', 'Cyclist']:
                continue
            h, w, l = float(obj[8]), float(obj[9]), float(obj[10])
            x, y, z = float(obj[11]), float(obj[12]), float(obj[13])
            box = {
                'type': obj[0],
                'dimensions': [h, w, l],
                'location': [x, y, z]
            }
            boxes.append(box)
    return boxes

# Visualize point cloud and boxes
def draw_pointcloud_with_boxes(points, boxes):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    canvas = FigureCanvas(fig)

    # Point cloud
    ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=0.1, c='gray')

    for box in boxes:
        h, w, l = box['dimensions']
        x, y, z = box['location']

        # Simple bounding box (not rotated)
        corners = np.array([
            [x - l/2, y - w/2, z],
            [x - l/2, y + w/2, z],
            [x + l/2, y + w/2, z],
            [x + l/2, y - w/2, z],
            [x - l/2, y - w/2, z + h],
            [x - l/2, y + w/2, z + h],
            [x + l/2, y + w/2, z + h],
            [x + l/2, y - w/2, z + h],
        ])

        edges = [
            [corners[i] for i in [0, 1, 2, 3]],  # bottom
            [corners[i] for i in [4, 5, 6, 7]],  # top
            [corners[i] for i in [0, 1, 5, 4]],
            [corners[i] for i in [1, 2, 6, 5]],
            [corners[i] for i in [2, 3, 7, 6]],
            [corners[i] for i in [3, 0, 4, 7]],
        ]

        ax.add_collection3d(Poly3DCollection(edges, facecolors='cyan', linewidths=0.5, edgecolors='r', alpha=0.3))

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_xlim(-30, 30)
    ax.set_ylim(-30, 30)
    ax.set_zlim(-3, 5)
    ax.view_init(elev=20, azim=-90)

    # Render to image
    canvas.draw()
    image = np.frombuffer(canvas.buffer_rgba(), dtype='uint8')
    image = image.reshape(canvas.get_width_height()[::-1] + (4,))
    plt.close(fig)

    return image[:, :, :3]  # drop alpha

# Paths
velodyne_dir = '/content/drive/MyDrive/kitti/training/velodyne'
label_dir = '/content/drive/MyDrive/kitti/training/label_2'
output_video_path = 'kitti_lidar_video.mp4' # .txt files

# Get sorted file list
frame_ids = sorted([f.split('.')[0] for f in os.listdir(velodyne_dir) if f.endswith('.bin')])
frames = []

# Process a few frames
for i, frame_id in enumerate(frame_ids[:10]):
    bin_path = os.path.join(velodyne_dir, f"{frame_id}.bin")
    label_path = os.path.join(label_dir, f"{frame_id}.txt")

    points = load_velodyne_bin(bin_path)
    boxes = load_kitti_labels(label_path)
    image = draw_pointcloud_with_boxes(points, boxes)
    frames.append(image)

# Save video
output_video_path = "kitti_3d_output.mp4"
imageio.mimsave(output_video_path, frames, fps=3)
print(f"✅ Video saved to: {output_video_path}")






```

```python
import re
import matplotlib.pyplot as plt

log_path = '/content/drive/MyDrive/mmdetection3d/work_dirs/pp_secfpn_80e/20250730_163133/20250730_163133.log'

# Metrics to extract
metrics = ['loss', 'loss_cls', 'loss_bbox', 'lr']
data = {k: [] for k in metrics}
iterations = []

with open(log_path, 'r') as f:
    for line in f:
        if 'mmengine' in line and 'Epoch(' in line and 'iter' in line:
            # Example line format:
            # "Epoch([1][20/1493])  lr: 0.001, loss: 1.2345, loss_cls: 0.456, loss_bbox: 0.789 ..."
            iter_match = re.search(r'\[(\d+)/\d+\]\)', line)
            if iter_match:
                iterations.append(int(iter_match.group(1)))

                for k in metrics:
                    match = re.search(rf'{k}:\s*([\d.]+)', line)
                    if match:
                        data[k].append(float(match.group(1)))
                    else:
                        data[k].append(None)  # In case metric is missing

# Plotting
plt.figure(figsize=(10, 6))
for k in metrics:
    if any(data[k]):
        plt.plot(iterations, data[k], label=k)

plt.xlabel('Iteration')
plt.ylabel('Metric Value')
plt.title('Training Metrics from Raw Log')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
